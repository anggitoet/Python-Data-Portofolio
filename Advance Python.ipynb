2.	Penulisan nama kolom menggunakan spasi tidak sesuai dengan kaidah penamaan kolom yang baik. Buat program untuk mengubah nama kolom sesuai dengan kaidah penulisan camel_case. camel_case merupakan kaidah penulisan dengan aturan :
    1. Setiap kata dipisahkan oleh simbol underscore_ 
    2. Setiap karakter wajib huruf kecil
    
    Tunjukkan kode yang anda buat beserta keluaran nama kolom yang baru

```python
def camel_case(s):
    return s.lower().replace(' ','_')
df.columns = [camel_case(col) for col in df.columns]
df
```

# Data Visualization
1.	Buatlah visualisasi untuk menunjukkan gross profit setiap quarternya.

```python
df.plot(x='date', y='gross_profit', kind='line', marker='o', color='green')
plt.grid(True)
```

2.	Dari plot nomor 1, coba buat suatu penanda bahwa pada pertengahan 2020 telah terjadi pandemi covid-19.

```python
df.plot(x='date', y='gross_profit', kind='line', marker='o', color='green')
plt.grid(True)

ax = plt.gca()

line_date = pd.to_datetime('2020-09-30')
event = "Covid-19"

ax.axvline(x=line_date, color='red', linestyle='--', linewidth=1)
ax.text(line_date, df['gross_profit'].max(), event, rotation=0, verticalalignment='bottom', color='red')
```
